```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Data
csv = pd.read_csv('Population/Population - Cleaned.csv')
```

```python
# Clean
pop = csv.drop(columns=['Unnamed: 0', '2020'])
#pop
```

```python
# Rename Columns
cols = ['Country', 1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 
        2013, 2014, 2015, 2016, 2017, 2018, 2019]
pop.columns = cols
#pop
```

```python
pop = pop.astype({1990: int, 1991: int, 1992: int, 1993: int, 1994: int, 1995: int, 1996: int, 1997: int, 1998: int, 1999: int, 2000: int, 2001: int, 2002: int,
                  2003: int, 2004: int, 2005: int, 2006: int, 2007: int, 2008: int, 2009: int, 2010: int, 2011: int, 2012: int, 2013: int, 2014: int, 2015: int,
                  2016: int, 2017: int, 2018: int, 2019: int})
#pop
```

```python
# Switch Columns and Rows
transpose = pop.T
#transpose
```

```python
transpose.columns = transpose.iloc[0]
#transpose
```

```python
transpose = transpose.drop(transpose.index[0])
#transpose
```

```python
transpose.reset_index(inplace=True)
```

```python
newDF = transpose.copy()
#newDF
```

```python
newDF = newDF.rename_axis(None, axis=1)
#newDF
```

```python
newDF = newDF.rename({'index' : 'Year', 'Russian Federation': 'Russia'}, axis='columns')
#newDF
```

```python
cleaned = newDF.astype(int)
cleaned
```

```python
# Australia
X = cleaned['Year'].values.reshape(-1, 1)
y = cleaned['Australia']

print('Australia: ', X.shape, y. shape)
```

```python
# Start Machine Learning
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=29)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Train Model
model.fit(X_train, y_train)
```

```python
# Score
model.score(X_test, y_test)
```

```python
# Australia: Predictions vs Actual
predictions = model.predict(X)
print(f"True output: {y[0]}")
print(f"Predicted output: {predictions[0]}")
print(f"Prediction Error: {predictions[0]-y[0]}")
```

```python
pd.DataFrame({"Predicted": predictions, "Actual": y, "Error": predictions - y})[["Predicted", "Actual", "Error"]]
```

```python
# Save Transposed Data
cleaned.to_csv('Population/Population_Transposed.csv')
```
